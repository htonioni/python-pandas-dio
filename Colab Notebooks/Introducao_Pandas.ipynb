```python
# Importando PANDAS com apelido de PNDA
import pandas as pnda
```

```python
# Definindo a variável dataframe (frame de dados)
dataframe = pnda.read_csv('/content/drive/MyDrive/datasets/Gapminder.csv', error_bad_lines=False, sep=";")
```

```python
# Mostra o conteúdo dessa lista
dataframe.head()
```

```python
# Altera as colunas dessa lista (utilizando dictionary)
dataframe = dataframe.rename(columns={'country':'País', 'continent':'Continente', 'year':'Ano', 'lifeExp':'Expct_de_Vida', 'pop':'População', 'gdpPercap':'Renda_per_Capta'})
```

```python
# Na coluna continente, retornar apenas valores únicos (a lista de países)
dataframe["Continente"].unique()
```

```python
# a variavel Americas_hugo recebe apenas as linhas do continente AMERICA
# Logo apos estamos dando display nessa variável
Americas_hugo = dataframe.loc[dataframe['Continente'] == 'Americas']
Americas_hugo.head()
```

```python
# Conta quantos paises há em cada continente
dataframe.groupby('Continente')['País'].nunique()
```

```python
# groupby agrupa o 'Ano' e imprime a coluna expectativa de vida 

dataframe.groupby('Ano')['Expct_de_Vida'].mean()
```
