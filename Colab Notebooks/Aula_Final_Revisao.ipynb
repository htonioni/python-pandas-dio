```python
# Qual a receita total? (soma de uma coluna total)
datafr['Valor Venda'].sum()
```

```python
# Criando nova coluna 'Custo Total', recebendo a multi de ['Custo Unit.' X 'Quantidade']
datafr['Custo Total'] = datafr['Custo Unitário'].mul(datafr['Quantidade'])
```

```python
datafr.head()
```

```python
# Soma  da coluna custo total arredondando DUAS casas
round(datafr['Custo Total'].sum(), 2)
```

```python
# Criando coluna de LUCRO, sendo Receita (valor de venda) -  Valor de Custo TT
datafr['Lucro'] = datafr['Valor Venda'] - datafr['Custo Total']
```

```python
datafr.sample(5)
```

```python
# Calculando lucro total
round(datafr['Lucro'].sum() , 2)
```

## 📚 Vamos calcular quanto tempo levou entre **Venda** e **Envio de produto**

```python
# Criando coluna tempo de envio
datafr['Tempo de envio'] = datafr['Data Envio'] - datafr['Data Venda']
```

```python
datafr.sample(2)
```

```python
# Média de tempo de envio para cada Marca, primeiro precisamos extrair: days
datafr['Tempo de envio'] = (datafr['Data Envio'] - datafr['Data Venda']).dt.days
```

```python
datafr.head(1)
```

```python
# Verificando se a coluna dtype passou para int
datafr['Tempo de envio'].dtype
```

```python
# Agora sim podemos calcular a media de tempo de envio POR MARCA
# Realizamos o agrupamento por "MARCA" (.groupby())
# Pego a coluna "TEMPO DE ENVIO" e faço .mean()[# que é a media]
# realizando o arredondamento com duas casas
round(datafr.groupby('Marca')['Tempo de envio'].mean(), 2)
```

```python
# Verificamos se temos dados faltantes
datafr.isnull().sum()
```

# Sabendo o **Lucro** por **Ano** e por **Marca**

```python
# Agrupar por 'Ano', depois por 'Marca' e somando o 'Lucro' de cada 'Marca'
# Tambem realizando arredondamento com 2 casas
datafr.groupby([datafr['Data Venda'].dt.year, 'Marca'])['Lucro'].sum()
```

```python
# Realizando a config de display dos resultados para virgula e 2 numeros apos a virgula
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
# Salvando em uma var e resetando o INDEX
lucro_ano = datafr.groupby([datafr['Data Venda'].dt.year, 'Marca'])['Lucro'].sum().reset_index()
lucro_ano
```

```python
# Qual a quantidade total de produtos vendidos ?
datafr['Quantidade'].sum()
```

```python
# Qual a 'Quantidade' vendida de produtos agrupada por 'Produto'
datafr.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
```

```python
# Passando estas informações para um gráfico, so adiconar o .plot
# Junto com o tipo de grafico('.bar()', '.barh()', etc)
# Adicionando tambem um Título
# E na linha X adicionando a label 'Total'
# Ponto e virgula no final para nao mostrar descrição
datafr.groupby('Produto')['Quantidade'].sum().sort_values(ascending=True).plot.barh(title='Total de produtos Vendidos');
plt.xlabel('Total');
```

```python
# Agrupando lucro por ano
#             Extraindo de 'Data Venda' o ano     e pegando a soma do lucro
datafr.groupby(datafr['Data Venda'].dt.year)['Lucro'].sum().plot.bar(title='Lucro x Ano');
plt.xlabel('Ano');
plt.ylabel('Lucro');
```

```python
# Sem grafico apenas para analise de numeros
datafr.groupby(datafr['Data Venda'].dt.year)['Lucro'].sum()
```

```python
# Selecionando apenas 2009
datafr_2009 = datafr[datafr['Data Venda'].dt.year == 2009]
```

```python
datafr_2009.head(4)
```

```python
# Criando gráfico de linha (usar o '.plot' vazio)
# Agrupado por mês de venda a qntd de lucro no ano de 2009 (var 2009)
datafr_2009.groupby(datafr_2009['Data Venda'].dt.month)['Lucro'].sum().plot(title='Lucro x Mes')
```

```python
# Lucro x Marca com rotação horizontal nas marcas ( x )
datafr_2009.groupby('Marca')['Lucro'].sum().plot.bar(title='Lucro x Marca')
plt.xlabel('Marca')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```
